使用量計算

```python
def Usage(data):
    for i in range(len(data)):
        target_1 = data.loc[i:i].reset_index(drop=True)
        target_2 = data.loc[i+1:i+1].reset_index(drop=True)
        if target_2.empty:
            break
        #         print(target_1['週'].values[0])
#         計算目標1和目標2相差幾週
        length = target_2['週'].values[0] - target_1['週'].values[0]
        if length == 1:
            data.loc[i,['使用量']] = [target_1['']]
            
        
        data.loc[i,['使用量']] = [11]
#     print(data)
```

```python
def label(df):
    if df['年'] == 2020:
        return df['週']+52
    elif df['年'] == 2021:
        return df['週']+104
    else:
        return df['週']

```

```python
# 衛材
part_no = 'B07109004'

data=pd.read_csv(f"./沒有紀錄的衛材領用量/五病房/{part_no}.csv")
data = data[data.columns.drop(list(data.filter(regex='Unnamed')))]
# 刪除第53週
data = data.drop(data[data['週'] == 53].index).reset_index(drop=True)
data['使用量'] = 0

# 把2020、2021年的週+52+104
# 加入年的欄位
data["日期"] = pd.to_datetime(data["日期"] ,format='%Y/%m/%d')
data['年'] = data['日期'].dt.isocalendar().year
data['週']=data.apply(label,axis=1)
# print(data)
Usage(data)
```
