```python
import pandas as pd
weather = pd.read_csv(f"./2022溫度差資料.csv",encoding='utf_8_sig')
weather = weather[weather.columns.drop(list(weather.filter(regex='Unnamed')))]
weather
```

```python
import pyodbc
SERVER = 'DESKTOP-OIEIPGU'
DATABASE = 'NCKUH'
conn = pyodbc.connect('DRIVER={ODBC Driver 17 for SQL Server};\
                       SERVER='+SERVER+';\
                       DATABASE='+DATABASE+';\
                       Trusted_Connection=yes;')
cursor = conn.cursor()
for index, row in weather.iterrows():
     cursor.execute("INSERT INTO weather (date,year,week,max_temperature,min_temperature,diff_temperature) values(?,?,?,?,?,?)", row.日期, row.年	, row.週,row.最高溫,row.最低溫,row.最大溫差)
conn.commit()
cursor.close()
```
