```python
import pandas as pd
import os
import numpy as np
```

幫合併完的資料增加日期的欄位

```python
date =  pd.DataFrame()
# date['日期'] = pd.DataFrame(pd.date_range(f'1/1/{2019}',f'31/12/{2021}'))
date['日期'] = pd.DataFrame(pd.date_range(f'1/1/{2019}',f'31/12/{2022}'))
date['週'] = date['日期'].dt.isocalendar().week
date['年'] = date['日期'].dt.isocalendar().year
# date = date.drop(date[date['週'] == 53].index).reset_index()
# del_data =  date[date['週'] == 52].tail(1).index
# date = date.drop(date[del_data.values[0]+1:].index).reset_index()
# date["日期"] = pd.to_datetime(date["日期"] ,format='%Y/%m/%d')
date 
```

讀取所有天氣資料檔案

```python
total_data = pd.DataFrame()
total_data = pd.read_csv(f'./Observation_CWB/C0K400.csv')
columns = ['Temperature','date']
total_data = total_data[columns]
total_data['Temperature'] = total_data['Temperature'].apply(lambda x: float(x))
total_data['date'] = pd.to_datetime(total_data['date'])
total_data
```

計算每天的溫度差

```python
answer = pd.DataFrame()
for i in range(len(date)):
    target = date.loc[i:i].reset_index(drop=True)
    temp = total_data[total_data['date'] == target['日期'].values[0]]
    # 最大值
    max_value = temp['Temperature'].max()
    # 最小值
    min_value = temp['Temperature'].min()
    Difference = max_value - min_value
    value = {"日期":target['日期'].values[0],'年':target['年'].values[0],'週':target['週'].values[0],"最高溫":max_value,"最低溫":min_value,
             "最大溫差": round(Difference, 2)}
    answer = answer.append(value,ignore_index=True)

    

    
answer
```

```python
# answer.loc[(answer['週']==1)&(answer['年']==2019),'平均'] = 1
answer.head(10)
```

```python
years = [2019,2020,2021]
for year in years:
    for i in range(1,53):
        temp = answer[(answer["年"] == year)&(answer["週"] == i)]
        temp_average = temp["最大溫差"].mean()
        temp_average = round(temp_average, 2)
        answer.loc[(answer['週']==i)&(answer['年']==year),'最大平均溫度差'] = temp_average
print(answer)
```

計算一週有幾最大溫差是Q3以上

```python
# Temperature['最大溫差'].mean()
# 只計算2019，2020
# Temperature = answer[(answer["年"] ==2019)|(answer["年"] ==2020)]
Temperature = answer[(answer["年"] ==2019)|(answer["年"] ==2020)|(answer["年"] ==2021)]
Q1=np.quantile(Temperature['最大溫差'],0.25,interpolation='lower')#下四分位数
# Q3=np.quantile(Y_true,0.75,interpolation='higher')#上四分位数
Q3 = np.percentile(Temperature['最大溫差'],75)
print(Q3)
# IQR = Q3 - Q1
# outlier = IQR*1.5

years = [2019,2020,2021]
for year in years:
    for i in range(1,53):
        temp = answer[(answer["年"] == year)&(answer["週"] == i)]
        count = temp[temp['最大溫差'] > Q3]
        
        answer.loc[(answer['週']==i)&(answer['年']==year),'異常溫差天數'] = len(count)
```
