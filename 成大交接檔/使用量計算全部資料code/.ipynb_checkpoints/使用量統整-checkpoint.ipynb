```python
import xgboost as xgb
import pandas as pd
import numpy as np
import sklearn
from chinese_calendar import is_workday
```

```python
# 先製造日期在計算工作日，接著在和原本資料合併
```

```python
def work_day(data,year):
    check_date = pd.read_csv(f"./週期資料.csv")
    answer =  pd.DataFrame()
    date =  pd.DataFrame()
    date['日期'] = pd.DataFrame(pd.date_range(f'1/1/{year}',f'31/12/{year}'))
    date['週'] = date['日期'].dt.isocalendar().week
    date = date.drop(date[date['週'] == 53].index).reset_index()
    del_data =  date[date['週'] == 52].tail(1).index
    date = date.drop(date[del_data.values[0]+1:].index).reset_index()
    date["日期"] = pd.to_datetime(date["日期"] ,format='%Y/%m/%d')
    check_date["date"] = pd.to_datetime(check_date["date"] ,format='%Y/%m/%d')
    # 如果是工作日就存成true，否則存成false
    date['休假日']=date['日期'].map(lambda x:(check_date['date']==x).any())
    print(date)

    for num in range(1, 53,1):
        temp = date[date['週'] == num]
#         計算該周工作日
        holiday = len(temp[temp['休假日'] == True])
        work_day = len(temp[temp['休假日'] == False])
        date.loc[date['週']==num,'工作天數'] = work_day
        date.loc[date['週']==num,'休假天數'] = holiday
        if year == 2021 and num == 52:   
            date.loc[date['週']==num,'休假天數'] = 2

#     同一週期工作日、休假日都一樣所以只保留一筆
    date.drop_duplicates(subset='週', keep='last', inplace=True)
    date = date.drop(['日期', '休假日','level_0','index'], axis=1)
    print(date)
    answer = pd.merge(data, date, on='週')
#     刪除日期和工作日

    
    return answer
```

每週資料量相加

```python
def week_data_sum(data):
    part_compute = pd.DataFrame()
    total_sum = 0
    for num in range(1, 53,1):
        temp = data[data['週'] == num]
        last_temp = temp.tail(1)
        if len(temp) == 0:
            last_temp['數量'] = 0
            part_compute = part_compute.append(last_temp)
        else:
            temp = temp.drop(temp[temp['數量'] < 0].index)
            last_temp['數量'] =  temp['數量'].sum()
            part_compute = part_compute.append(last_temp)
#         total_sum = total_sum +part_compute['數量'].values[0]
    
    return part_compute
```

讀取資料+整理工作日資料

```python
# 料號
# A00120033
# A00120067
# A00120110
# A00120213
# A00421190
# A02120180
# A02320340
# A02322259
# A03295054
# A04200004
# A04411285
# A04411370
# A04800036
# B00206057
# B00304056
# B03106859
# B03110120
# B08143445
# C02113123
# C13830005
# C13930399
# G81500272
# K80004044

# total_part_no = ['A00120033','A00120067','A00120110','A00120213','A00421190','A02120180','A02320340','A02322259','A03295054','A04200004','A04411285','A04411370','A04800036'
#            ,'B00206057','B00304056','B03106859','B03110120','B08143445','C02113123','C13830005','C13930399','G81500272','K80004044']
total_part_no = ['B08143445']
# 病房
room = '五病房'
# 年分
years = [2019,2020,2021]

total_data =  pd.DataFrame()

for part_no in total_part_no:
    for year in years:
        data = pd.read_csv(f"../使用量計算/全部資料合併/{year}_{room}_合併檔.csv",encoding='utf_8_sig')
        data = data[data.columns.drop(list(data.filter(regex='Unnamed')))].reset_index(drop=True)
        data = data[data['料號'] == part_no]
        data = data.drop(data[data['數量'] < 0].index).reset_index()
        # 先刪除大於52周的資料
        del_data =  data[data['週'] == 52].tail(1).index
        if(len(del_data) !=0 ):
            data = data.drop(data[del_data.values[0]+1:].index).reset_index()
        # 計算工作日
        data = work_day(data,year)
        total_data = pd.concat([total_data, data])
        #     刪除多的欄位
    total_data = total_data[total_data.columns.drop(list(total_data.filter(regex='index')))].reset_index(drop=True)
    total_data = total_data[total_data.columns.drop(list(total_data.filter(regex='level_0')))].reset_index(drop=True)
    # total_data.to_csv(f"./有加工作日/{room}/{part_no}.csv",encoding='utf_8_sig')

    # 把每一週的使用量加總
    total_data['年'] =  pd.to_datetime(total_data['帳務日期']).dt.year
    data_2019 = total_data[total_data['年']==2019]
    data_2020 = total_data[total_data['年']==2020]
    data_2021 = total_data[total_data['年']==2021]
    data_2019 = week_data_sum(data_2019)
    data_2020 = week_data_sum(data_2020)
    data_2021 = week_data_sum(data_2021)
    total_data = pd.concat([data_2019, data_2020])
    total_data = pd.concat([total_data, data_2021])
    total_data.to_csv(f"./衛材相關性計算/{room}/{part_no}.csv",encoding='utf_8_sig')
    
```
