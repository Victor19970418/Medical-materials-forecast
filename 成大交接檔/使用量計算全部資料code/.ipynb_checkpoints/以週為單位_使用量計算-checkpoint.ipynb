針對該料號計算每年每週使用量

```python
# 料號
# A00120067
# A00120110
# A02120180
# A02320340
# A00120213
# 年分
part_no = "B00304056"
years = [2019,2020,2021]
for year in years:

    #先讀取合併檔，計算每週使用量
    data_108=pd.read_csv(f"./全部資料合併/{year}_五病房_合併檔.csv")

    # 依料號資料篩選
    data_108 = data_108[data_108['料號'] == part_no]

    #compute
    part_compute = pd.DataFrame()
    # 需要儲存的欄位
    columns = ['庫房','料號','品名','數量','週']

    total_sum = 0
    for num in range(1, 53,1):
        temp = data_108[data_108['週'] == num]
        last_temp = temp.tail(1)[columns]
        if len(temp) == 0:
            last_temp['數量'] = 0
            part_compute = part_compute.append(last_temp)
        else:
            temp = temp.drop(temp[temp['數量'] < 0].index)
            last_temp['數量'] =  temp['數量'].sum()
            part_compute = part_compute.append(last_temp)
        total_sum = total_sum +part_compute['數量'].values[0]
    print(part_compute)
    print(total_sum)
    part_compute.to_csv(f"./以週為單位/{year}/{part_compute['料號'].values[0]}_{year}週期資料.csv",encoding='utf_8_sig') 
```
