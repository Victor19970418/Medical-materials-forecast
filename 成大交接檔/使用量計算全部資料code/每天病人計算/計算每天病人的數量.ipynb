```python
import pandas as pd
import os
```

```python
# 讀取兩份資料，一年分成兩份
data_1=pd.read_csv(f'./刪除重複值/五病房_5.csv')
data_2=pd.read_csv(f'./刪除重複值/五病房_6.csv')
total_data = pd.concat([data_1, data_2])
total_data = total_data[total_data.columns.drop(list(total_data.filter(regex='Unnamed')))]
```

產生一年的日期資料

```python
year = 2021
date =  pd.DataFrame()
date['日期'] = pd.DataFrame(pd.date_range(f'1/1/{year}',f'31/12/{year}'))
date
```

計算每天有幾個病床紀錄

```python
date['日期'] = pd.to_datetime(date['日期'])
date['日期']=  date['日期'].dt.date
total_data['date'] = pd.to_datetime(total_data['date'])
total_data['date']=  total_data['date'].dt.date
for i in range(len(date)):
    target_day = date.loc[i:i].reset_index(drop=True)
    temp = total_data[total_data['date'] == target_day['日期'].values[0]]
    print(len(temp))
    date.loc[i:i,'病人數量'] = len(temp)
```
